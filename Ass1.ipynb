```python
# Adding feature
import pandas as pd
data = pd.read_csv("used_cars_data.csv")
from datetime import date
date.today().year
data['Car_Age']=date.today().year - data['Year']
data.head()
```

```python
# Replace NaNs with median value of the column
data['New_Price'] = data['New_Price'].fillna(data['New_Price'].median())

data['Profit'] = data['New_Price'] - data['Price']
data['Avg_kilo_year'] = round(data['Kilometers_Driven'] / data['Car_Age'])
data.head()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from a CSV file (update the path as needed)
df = pd.read_csv('used_cars_data.csv')  # Replace 'your_file.csv' with the actual filename

# Encode categorical variables
df_encoded = df.copy()
df_encoded['Fuel_Type_Encoded'] = df['Fuel_Type'].astype('category').cat.codes
df_encoded['Name_Encoded'] = df['Name'].astype('category').cat.codes

# Plot
plt.figure(figsize=(6,4))
sns.scatterplot(data=df_encoded, x='Fuel_Type_Encoded', y='Name_Encoded', hue='Fuel_Type', palette='Set2', alpha=0.6)

plt.title('Fuel Type vs Car Model')
plt.xlabel('Fuel Type (Encoded)')
plt.ylabel('Model (Encoded)')
plt.legend(title='Fuel Type')
plt.grid(True)
plt.tight_layout()
plt.show()
```
