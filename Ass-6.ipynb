```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

df=pd.read_csv("diabetes.csv")
col_names = df.columns.tolist()
print(col_names)
df.head()
```

```python

feature_col= ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X=df[feature_col]
y=df['Outcome']

```

```python
# Splitting data 
X_train,X_test,y_train,y_test = train_test_split(X ,y , test_size=0.30,random_state=1)

# decision tree
clf=DecisionTreeClassifier()
clf=clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn import tree


pima = pd.read_csv("diabetes.csv")


print("Columns in dataset:", pima.columns.tolist())


feature_cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness',
                'Insulin','BMI','DiabetesPedigreeFunction','Age']

X = pima[feature_cols]     
y = pima['Outcome']        


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)


clf = DecisionTreeClassifier()


clf = clf.fit(X_train, y_train)


y_pred = clf.predict(X_test)


print("Accuracy:", metrics.accuracy_score(y_test, y_pred))


plt.figure(figsize=(15,10))
tree.plot_tree(clf,
               feature_names=feature_cols,
               class_names=['No Diabetes','Diabetes'],
               filled=True,
               rounded=True)
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn import tree


pima = pd.read_csv("diabetes.csv")


feature_cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness',
                'Insulin','BMI','DiabetesPedigreeFunction','Age']

X = pima[feature_cols]
y = pima['Outcome']


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)


clf_full = DecisionTreeClassifier(random_state=1)
clf_full = clf_full.fit(X_train, y_train)

plt.figure(figsize=(15,10))
tree.plot_tree(clf_full,
               feature_names=feature_cols,
               class_names=['No Diabetes','Diabetes'],
               filled=True, rounded=True)
plt.title("Full Decision Tree")
plt.show()


clf_4 = DecisionTreeClassifier(max_depth=4, random_state=1)
clf_4 = clf_4.fit(X_train, y_train)

plt.figure(figsize=(15,10))
tree.plot_tree(clf_4,
               feature_names=feature_cols,
               class_names=['No Diabetes','Diabetes'],
               filled=True, rounded=True)
plt.title("Decision Tree (Max Depth = 4)")
plt.show()


y_pred_4 = clf_4.predict(X_test)
print("Accuracy with max_depth=4:", metrics.accuracy_score(y_test, y_pred_4))
```
